## The set_index and reset_index Methods
- The index serves as the collection of primary identifiers/labels/entrypoints for the rows.
- The fastest way to extract a row is from a sorted index by position/label
- Pandas uses index labels/values when merging different objects together.
- The `set_index` method sets an existing column as the index of the **DataFrame**
- The `reset_index` method sets the standard ascending numeric index as the index of the df
- set and reset are not mutating


```python
bond = bond_df.copy()
bond.head()
```

```python
bond = bond.set_index("Film")
```

```python
bond
```

```python
bond.reset_index()
```

```python
bond = bond_df.copy()
bond.set_index("Film").set_index("Year") # Setting consecutavely will kill old index... useful to reset index between sets
```

## Retrieve Rows by Index Label with loc Accessor
- The loc accessor retrieves one or more rows by index label
- Provide a pair of square brackets after the accessor

```python
bond = bond_df.copy().set_index("Film").sort_index() # Sort index helps speed of lookups
bond.head()
```

```python
bond.loc["Dr. No"]
```

```python
bond.loc[["From Russia with Love", "Goldfinger"]]
```

```python
bond.loc["Dimonds Are Forever": "Moonraker"] # loc accessor is inclusive of last elem unlike iloc
```

```python
bond.iloc[0:4] # this pulls 0, 1, 2, 3
```

## Overwrite Multiple Values in a Data Frame
- The replace method replaces all occurences of a **Series** value with another value (think of it like "Find and Replace").
- To overwrite multiple values on a **DF** remember to use an accessor on the df
- `loc` and `iloc` can accept a boolean series. Use them to target the values to overwrite

```python
bond = bond_df.copy().set_index("Film").sort_index()
bond.head()
```

```python
bond["Actor"] = bond["Actor"].replace("Sean Connery", "Sir Sean Connery")
```

```python
bond
```

```python
bond = bond_df.copy().set_index("Film").sort_index()
bond.head()
```

```python
bond[bond["Actor"] == "Sean Connery"].loc[:, "Actor"] = "Sir Sean Connery"
```

```python
bond.loc[bond["Actor"] == "Sean Connery", "Actor"] = "Sir Sean Connery"
```

```python
bond
```

## Delete Rows or Columns from a Dataframe
- `drop` method will delete rows/columns from the df non-mutating
- pass `index` or `columns` param with column names to remove
- The `pop` column removes and returns a single **Series** mutating df
- Python `del` method also removes a column


```python
bond = bond_df.copy().set_index("Film").sort_index()
bond.head()
```

```python
bond.drop(columns=["Box Office", "Budget"])
```

```python
bond
```

```python
bond.drop(index=["No Time to Die"]) # If this index occurs +1 times then will drop all of those
```

```python
bond.drop(index=["No Time to Die", "Casino Royale"], columns=["Box Office", "Budget"])
```

## The nsmallest and nlargest Methods
- The `nlargest` method returns a specified number of rows with the largest values from a given column
- The `nsmallest` method returns rows with the smallest values from a given column.
- the `nlargest` and `nsmallest` methods are more efficient than sorting the entire `DataFrame`.

```python
bond = bond_df.copy().set_index("Film").sort_index()
bond.head()
```

```python
bond.nsmallest(3, columns=["Year", "Box Office"])
```

```python
bond.nlargest(4, columns=["Box Office"])
```

```python
bond["Box Office"].nlargest(4)
```

## Filtering with the where Method
- Similar to square brackets or `loc`, the `where` method filters the original `DataFrame` with a Boolean Series.
- All rows present but rows that have `False` in the boolean series will have `NaN` populate for every cell value

```python
bond = bond_df.copy().set_index("Film").sort_index()
bond.head()
```

```python
bond.where(bond["Actor"] == "Sean Connery")
```

## The apply Method w/ DF

```python
bond = bond_df.copy().set_index("Film").sort_index()
bond.head()
```

```python
bond.apply(lambda row: print(row.loc["Year"]), axis="columns")
```

```python
def double_budget(row):
    new_row = row.copy()
    new_row.loc["Budget"] = new_row.loc["Budget"]  * 2
    return new_row
```

```python
bond.apply(double_budget, axis="columns")
```
